```python
import csv
import random

# Open the CSV file for reading
with open('Trips.csv', 'r') as csv_file:
    # Initialize a CSV reader
    csv_reader = csv.reader(csv_file)

    # Skip the header row if there is one
    next(csv_reader, None)

    # Read all the numbers from the CSV file into a list
    numbers = [int(row[0]) for row in csv_reader]

# Ensure you have at least 100 numbers
if len(numbers) < 100:
    print("Not enough numbers in the CSV file.")
else:
    # Randomly sample 100 numbers from the list
    sample = random.sample(numbers, 100)

    # Print or use the sample as needed
    for number in sample:
        print(number)
```

```python
mean = sum(sample) / len(sample)

print("Mean:", mean)
```
